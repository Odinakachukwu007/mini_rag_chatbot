```python
import pandas as pd

# Load the data
df = pd.read_csv("Data/general_conference_talks.csv")

# Preview a few rows
df.head()
```

```python
!pip install tiktoken
```

```python
import tiktoken

# Initialize tokenizer
tokenizer = tiktoken.get_encoding("cl100k_base")

# Function to split text into ~300-token chunks
def chunk_text(text, max_tokens=300):
    tokens = tokenizer.encode(text)
    chunks = []

    for i in range(0, len(tokens), max_tokens):
        chunk = tokens[i:i + max_tokens]
        decoded = tokenizer.decode(chunk)
        chunks.append(decoded)

    return chunks
```

```python
# Create new DataFrame with chunks
chunk_data = []

for index, row in df.iterrows():
    chunks = chunk_text(row["text"])  # 'text' column contains the content of the talk
    for chunk in chunks:
        chunk_data.append({
            "title": row["title"],
            "speaker": row["speaker"],
            "url": row["url"],
            "chunk": chunk
        })

chunk_df = pd.DataFrame(chunk_data)
chunk_df.to_csv("Data/chunked_conference_talks.csv", index=False)

print("✅ Chunking completed and saved!")
chunk_df.head()
```
